```python
def employee_management_system():
    # Step 1: Initialize the dictionary with sample employee data
    employee_data = {
        101: {'name': 'Satya', 'age': 27, 'department': 'HR', 'salary': 50000},
        102: {'name': 'Priya', 'age': 32, 'department': 'Engineering', 'salary': 65000},
        103: {'name': 'Alex', 'age': 29, 'department': 'Marketing', 'salary': 55000}
    }
    
    while True:
        # Step 2: Define the menu system
        print("\n===== EMPLOYEE MANAGEMENT SYSTEM =====")
        print("1. Add Employee")
        print("2. View All Employees")
        print("3. Search for Employee")
        print("4. Exit")
        
        choice = input("\nEnter your choice (1-4): ")
        
        if choice == '1':
            # Step 3: Add Employee Functionality
            print("\n----- ADD EMPLOYEE -----")
            
            # Get employee ID and validate it's unique
            while True:
                try:
                    emp_id = int(input("Enter Employee ID: "))
                    if emp_id in employee_data:
                        print("Employee ID already exists. Please enter a unique ID.")
                    else:
                        break
                except ValueError:
                    print("Please enter a valid numeric ID.")
            
            # Get other employee details
            name = input("Enter Employee Name: ")
            
            # Validate age is a number
            while True:
                try:
                    age = int(input("Enter Employee Age: "))
                    if age <= 0:
                        print("Age must be a positive number.")
                    else:
                        break
                except ValueError:
                    print("Please enter a valid age.")
            
            department = input("Enter Employee Department: ")
            
            # Validate salary is a number
            while True:
                try:
                    salary = float(input("Enter Employee Salary: "))
                    if salary < 0:
                        print("Salary cannot be negative.")
                    else:
                        break
                except ValueError:
                    print("Please enter a valid salary amount.")
            
            # Add employee to dictionary
            employee_data[emp_id] = {
                'name': name,
                'age': age,
                'department': department,
                'salary': salary
            }
            
            print(f"\nEmployee with ID {emp_id} has been successfully added!")
            
        elif choice == '2':
            # Step 4: View All Employees
            print("\n----- ALL EMPLOYEES -----")
            
            if not employee_data:
                print("No employees available.")
            else:
                # Print header
                print(f"{'ID':<8} {'Name':<15} {'Age':<8} {'Department':<15} {'Salary':<10}")
                print("-" * 60)
                
                # Print each employee's details
                for emp_id, details in employee_data.items():
                    print(f"{emp_id:<8} {details['name']:<15} {details['age']:<8} {details['department']:<15} ${details['salary']:<10,.2f}")
        
        elif choice == '3':
            # Step 5: Search for an Employee by ID
            print("\n----- SEARCH EMPLOYEE -----")
            
            try:
                search_id = int(input("Enter Employee ID to search: "))
                
                if search_id in employee_data:
                    details = employee_data[search_id]
                    print("\nEmployee Found!")
                    print(f"ID: {search_id}")
                    print(f"Name: {details['name']}")
                    print(f"Age: {details['age']}")
                    print(f"Department: {details['department']}")
                    print(f"Salary: ${details['salary']:,.2f}")
                else:
                    print("Employee not found.")
            except ValueError:
                print("Please enter a valid numeric ID.")
        
        elif choice == '4':
            # Step 6: Exit the Program
            print("\nThank you for using the Employee Management System. Goodbye!")
            break
        
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

# Run the program
if __name__ == "__main__":
    employee_management_system()
```
